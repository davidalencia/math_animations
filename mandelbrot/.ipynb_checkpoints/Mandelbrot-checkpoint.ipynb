# Mandelbrot

### Dimensions

```julia
const height = 600 ::Int64
const width = 400 ::Int64
# Maximum number of iterations for each point on the complex plane
const maxiterations = 100 ::Int64
```

```julia
function dist(x1, y1, x2, y2)::Float64
    x = x2 - x1
    y = y2 - y1
    return sqrt(x^2+y^2)
end
```

```julia
function numberMapping(n, start1, stop1, start2, stop2) ::Float64
  return ((n-start1)/(stop1-start1))*(stop2-start2)+start2;
end
```

```julia
function mandelbrotPoint(x::Float64, y::Float64)::Float64
    #Now we test, as we iterate z = z^2 + cm does z tend towards infinity?
    a = x;
    b = y;
    n = 0;
    while n < maxiterations
        aa = a^2;
        bb = b^2;
        twoab = 2.0 * a * b;
        a = aa - bb + x;
        b = twoab + y;
        #Infinty in our finite world is simple, let's just consider it 16
        if dist(aa, bb, 0.0, 0.0) > 16
            break;
        end
        n +=1
    end 
    return n
end
```

```julia
function mandelbrot_it_2_color(n) :: RGB{N0f8}
    if n==maxiterations
        return RGB{N0f8}(0,0,0)
    end
    hsvColorspace = numberMapping(n, 0, maxiterations, 0, 255);
    return convert(RGB{N0f8}, HSV{Float64}(hsvColorspace, 1, 0.9))
end
```

```julia
function mandelbrot(w, offsetx, offsety)
    function coors_2_ix(i, j)::Int64
        ix = ((i-1)+(j-1)*(width))+1 ::Int64
        return ix
    end

    h = w * height / width 

    # Start at negative half the width and height
    xmin = -w/2;
    ymin = -h/2 ;

    # x goes from xmin to xmax
    xmax = xmin + w;
    # y goes from ymin to ymax
    ymax = ymin + h;

    # Calculate amount we increment x,y for each pixel
    dx = (xmax - xmin) / (width);
    dy = (ymax - ymin) / (height);
    pixels = Array{RGB{N0f8},2}(undef, width, height)
    
    y = ymin +offsety  
    for j in 1:height
        x = xmin + offsetx
        for i in 1:width
            n = mandelbrotPoint(x,y)
            pix = coors_2_ix(i, j)
            pixels[pix]=mandelbrot_it_2_color(n)
            x += dx;
        end
        y += dy
    end
    return pixels
end
```

```julia
desde = 2
dx = 0.001
rango = (4 .^ (range(0.8,stop=-4.4,length=300)))
p = Progress(length(rango))
maxiterations = 100 ::Int64
print("mandelbrot from $desde to $dx")
framestack = map(function(v)
        y = numberMapping(v[2], 4, 0, -1, -1.86214) # 195, 20
        pixels = mandelbrot(v[2],  y, 0)
        global maxiterations
        maxiterations += 2
        next!(p)
        return pixels
    end, enumerate(rango)) #vector of 2D arrays

encoder_options = (crf=23, preset="medium")
framerate=24
print("render video")
open_video_out("video.mp4", framestack[1], framerate=framerate, encoder_options=encoder_options) do writer
    @showprogress for frame in framestack
        write(writer, frame)
    end
end
```
